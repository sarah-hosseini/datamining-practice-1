L) Remove all outliers with respect to IQR (Inter Quartile Range) and plot
cleared data

```python
cols = ['Income'] # The columns you want to search for outliers in

# Calculate quantiles and IQR
Q1 = df[cols].quantile(0.25) 
Q3 = df[cols].quantile(0.75)
IQR = Q3 - Q1

# Return a boolean array of the rows with (any) non-outlier column values
condition = ~((df[cols] < (Q1 - 1.5 * IQR)) | (df[cols] > (Q3 + 1.5 * IQR))).any(axis=1)

# Filter our dataframe based on condition
df = df[condition]
df.Income.plot.hist()
```

```python
x = df.Expenses
y = df.Income
p = np.poly1d(np.polyfit(x, y, 7))
p10 = np.poly1d(np.polyfit(x, y, 10))
print(p)

import matplotlib.pyplot as plt
xp = np.linspace(0, 2500, 100)
_ = plt.plot(x, y, '.', color='pink')
_ = plt.plot(xp, p(xp), '-', color='green')
_ = plt.plot(xp, p10(xp), '--', color='blue')
plt.show()
```

As seen above, a degree higher than 7 results in an unstable curve, therefore weaccept p with degree 7.
